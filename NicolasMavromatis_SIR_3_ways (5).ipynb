Here are the rates of change, repeated to better understand how each depends on a:

Sprime=-a*S*I,
Iprime=(a*S*I)-(b*I),
Rprime=b*I

The transmissibility constant is defined as the average number of contacts with infectious individuals per susceptible person daily (p) times the fraction of contacts that lead to new infections (q).
So a higher p could result from more averge daily contacts between susceptible and infected due to large social events such as sporting events where people are in close proximity.
A lower p could result from less contact between infected and susceptible such as because of quarantine and social distancing/isolation. 
A higher q could result from less wearing of masks and a lower q could result from more mask wearing.
A higher p and or q means a higher transmissibility constant, a, and a lower p and or q means a lower a. 

A higher transmissibility constant means more infection occurs, while a lower transmissibility constant means less infection occurs.

With a higher a, Sprime becomes greater, meaning a larger negative change in susceptible and a smaller number of susceptible, S, at each time since there is more transfer from susceptible to infected.

A higher a also means a larger rate of infection (Iprime) and therefore a larger number of infected (I) at each time due to a greater transfer from susceptible to infected.

This would mean a larger rate of recovery (Rprime) and a larger number of recovered (R) at each time due to a larger I. With a larger number of infected, the rate of recovery goes up and so does the number of recovered at each time, especially because the recovery coefficient is constant. 

A lower transmissibility constant has the opposite effect as above. This would result in a smaller rate of change from susceptible to infected (Sprime), more susceptible people (S), a smaller rate of infection (Iprime), less infected people (I) and therefore a lower rate of recovery (Rprime) and less people that recover (R) overall at each time since the recovery coefficient remains constant.  

A lower transmissibility constant also leads to faster plateaus, or flattening the curve. 

These changes are reflected in the two graphs, above, with the first graph corresponding to a larger transmissibility constant and the second corresponding to a smaller transmissibility constant.

```python
#HERE IS THE ORIGINAL DIFFERENTIAL CODE BLOCK, UNALTERED

#Using Differential Equations


# Total population, N.
N = 1000

timeperiod =160
# Initial number of infected and recovered individuals, I0 and R0.
I0=1
R0 =1
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0

D0 =0


#TRY modifying a and b

# Contact rate, a, and mean recovery rate, b, (in 1/days).

days=10 #(days to recover - how long are the infected able to pass it on )

a=.0005 #(more contact, how quickly it spreads)

b=1/days 

deathrate = .01
recover = 1 - deathrate





# A grid of time points (in days)
t = np.linspace(0, 160, timeperiod)

# The SIR model differential equations.
def deriv(y, t, N, a, b):
    S, I, R, D = y
    dSdt = -a * S * I 
    dIdt = a * S * I  - b * I
    dRdt = recover * (b * I)
    dDrt = deathrate * (b * I) 
    return dSdt, dIdt, dRdt, dDrt

# Initial conditions vector
y0 = S0, I0, R0, D0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, a, b))
S, I, R, D = ret.T

# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w')
#ax = fig.add_subplot(111, axis_bgcolor='#dddddd', axisbelow=True)
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, S/1000, 'b', alpha=0.5, lw=2, label='Susceptible')
ax.plot(t, I/1000, 'r', alpha=0.5, lw=2, label='Infected')
ax.plot(t, R/1000, 'g', alpha=0.5, lw=2, label='Recovered with immunity')
ax.plot(t, D/1000, 'y', alpha=0.5, lw=2, label='died')
ax.set_xlabel('Time /days')
ax.set_ylabel('Number (1000s)')
ax.set_ylim(0,1.2)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
plt.show()
```

The differential model is very similar to the iterative SIR model, above.
In particular, the mathematical basis for calculating Sprime, Rprime, and Iprime are mostly the same. The context for this was explained in comments right beneath question 1. 

One thing that is different is that S0 is calculated from the total population, N, minus I0 minus R0. That is, the number of susceptible is everybody in the total population that is not infected or recovered.

Another difference is that this model also includes a deathrate-where 99 percent recover and 1 percent die. So Rprime and Dprime are calculated by multiplying (bI) by each percent, respectively.

The main differences between the two models are in how they are implemented.

Here, t = np.linspace(0, 160, timeperiod) is used to make a time grid array from 0 to the end time, broken up into (timeperiod) sections, to set the x axis and set the time step for each derivative.

The time step before was calculated as deltat=(tfinal-tinitial)/numberofsteps.

Another difference is that this implementation defines a derivative function, which modifies a vector containing S,I,R,D as y.

It calculates each prime value (derivative) from the y that is fed into this function then returns these derivatives.

OdeInt stands for ordinary differential equation integration.

It then uses the odeInt function to integrate the SIRD equations over the time grid, t (or solve the differential equations).

This function returns a matrix which has k columns, one for each of the k variables, and a row for each time point specified in t. 

It then feeds the appropriate values into the corresponding vectors as S, I, R, D = ret.T.

Finally, it graphs t as the x and each of these as the y for four plots, but another difference is the scale is in 1000's of people.

The original SIR implementation worked with a for loop, from k=1 to numberofsteps. In each iteration, it calculated Sprime, Iprime, and Rprime from the previous values of S,I,R, then found the delta (change) in each by multiplying each derivative by the change in time. It then added these to the previous values of t, S,I,R to find the new values of each for the next iteration. This worked differently by appending each value of S,I,R,t to a different array then using these to graph. 

Overall, the effects of the two implementations are the same, but they are handled differently-one is solved iteratively while the other is solved with a differential equation integrater.

**SIR with Matrices**

**Question 6:**
How is this implementation the same and different than SIRPLOT and the differential equation SIRmodel? Can you explore changes in the transmission coefficent and recovery rate the same way? Answer here.

The matrix implementation is very different, and is derived from the linear algebra book.

In the matrix implementation, the rate of change of each value (S,I,R,D-Prime) involves a flat percentage from day to day.

By default, there is 5% new infection, 95% that remain susceptible, 4% that recover with no immunity, 10% that recover with immunity, and 1% infected that die each day.

Also, xt is a 4 vector, composed of the fractions (not absolute values) of S,I,R,D at time t in that order. So, the fractions for each day add up to 1.

The 4 vector for the next day, x(t+1) is calculated by multiplying the values of the previous day(x(t)) by a fixed matrix, A, consisting of these percent changes in the correct places (see matrix below).

Here is how the matrix vector multiplication works to find each x(t+1) set of fractions:

To find S(t+1) of the next day, you add the susceptible fraction from the previous day that did not become infected (.95*S(t)) to the infected fraction that recovered without immunity (.04*I(t)).

To find I(t+1) of the next day, you add the fraction of new people that become infected from the previous day (.05*S(t)) to the fraction that are infected and remain infected (.85*I(t)). The percent that remain infected is 85% because 4% recover without immunity, 10% recover with immunity, and 1% die each day. 

To find R(t+1) of the next day, you add the fraction that recovered with immunity from the previous day (.10*I(t)) to the fraction that remain recovered (1*R(t)).

Finally, to find D(t+1) of the next day, you add the fraction of infected that die (.01*I(t)) to the fraction already dead (1*D(t)). 

This is very different from the iterative and differential models which both relied on calculus, derivatives, and delta values. The original two models calculated the rates of change for each variable, multiplied by the change in time to find the delta values, then added these delta values onto the original values to find the values for the next day. The matrix model works differently by finding the values for the next day by multiplying those from the previous day by a flat percent and summing these terms. In a way, it combines the steps of calculating rates of change and delta values and adding these to the values of the previous day by using the sum of flat percentages. 

Another difference is the matrix model assumes some percent recover with no immunity each day.

Also, the other two models calculated the numerical values of S,I,R,D where the matrix model just calculates the relative fraction of each at each time, and populates a 4 vector xt. 

The implementations are also very different, with this model using successive matrix multiplication to model the spread of disease.

The models have some similarities. Both effectively calculate the rate of change and values of S,I,R,D for the next day from the values of S,I,R,D from the previous day.  



The original two models had a recovery coefficient, b, and a transmission coefficient, a, defined for each disease. The matrix model handles this differently where b roughly corresponds to the percent that recover with immunity each day. The transmission coefficient, a, roughly corresponds to the percent of new infection each day. 

To find the next S, rather than multiplying S and I by a, as in the first two models, S is multiplied by a flat percentage, I is multiplied by a flat percentage, and these values are summed. 

To find the next R, rather than multiplying I by b, it multiplies I by a flat percentage, which is effectively the same.

To find the next D, rather than Multiplying I by b and the percent that die, it multiplies I by a flat percentage each day. 

To find the next I, the calculation is very different. Rather than multiplying different terms by a and b and finding their difference, it adds the flat percent of S that become infected to the flat percent of I that remain infected. 

The recovery coefficient, b, is the average number of people that recover each day. This can be modelled here by changing the percent that recover with immunity each day.

The effects can be considered per unit time passed. As before, a higher number of people that recover with immunity means a lower number that remain infected. This means there are fewer people that are susceptible as more recover. This also leads to fewer deaths. 

Susceptibility is affected both by the percent that remain susceptible but also by the percent that recover without immunity. A higher percent that recover without immunity means a higher number that are susceptible.

The transmission coefficient, a, is a combination of the fraction of all possible contacts between S and I that occur, times the fraction of all contacts that lead to infection, with a higher a value meaning greater infection. This can be modelled here by changing the percent of new infection each day. The effects can be considered per unit time passed. A higher percent that become infected means a lower percent that remain susceptible. More infection means a higher number that recover. This also leads to more deaths. 

If you lower either of these two percents, the opposite happens as described above.

```python
#ORIGINAL CODE with more comments

T  = 210 #number of days
a= .05         # 5 % new infection each day
aS= 1 - a      # 95 % remain suseptible each day
rnI= .04       # 4% recover no immunity each day
rI= .1         # 10 % recover with immunity each day
rD= .01        # 1% die from disease each day


A = np.matrix([[aS,rnI,0,0],[rnI+rD,(1-(rnI +rI + rD)) ,0,0],[0,rI,1,0],[0,rD,0,1]]) #dynamics matrix
x1 = np.array([1,0,0,0]) #initial state: everyone healthy
stateTraj = np.hstack([np.vstack(x1),np.zeros((4,T-1))]) #initialize trajectory with 0s
for t in range(T-1):
    stateTraj[:,t+1] = np.matmul(A,stateTraj[:,t])

for i in range(len(stateTraj)):
    plt.plot(range(T),stateTraj[i])

plt.legend(["Susceptible","Infected","Recovered","Deceased"])
plt.title("Original Matrix Model Conditions")
```

The Lotka-Voerra Model models how the populations of lynxes and hares affect eachother. Lynxes feed on hares.

The hares experience unbounded population growth in this model, with constant per capita growth. This differs from the other model in the chapter by this simplification. 

The number of hares declines at a rate proprotional to HL, the number of hares and the number of Lynxes, respectively. This assumes rabbits do not die of old age.

In the absence of hares, the lynxes die off at a rate proportional to the number of lynxes. This is because, with more lynxes there is a higher population competing for limited resources, leading to starvation and death.

The lynx population increases at a rate proportional to HL, which roughly models the number of interactions between the groups. If lynxes can feed on more hares, they are able to reproduce at a higher rate which leads to lynx population growth.

This model relies on four constant parameters that model the interactions, population growth, and death between the two groups.

This was illustrated by the sine like patterns of the two lines that varied with one another, almost as if one was the x-shift of the other.

I think the syntax of the differential model made it a little easy to lose track of the ideas, as a lot happens under the hood with the built in algorithm that makes it harder to understand the basic calculus concepts of things like rates of change and delta values.
